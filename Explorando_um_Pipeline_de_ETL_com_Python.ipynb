**Transform**

```python
Data_Client['Message'] = list(map(generate_new_custom_news,Data_Client['ID']))
Data_Client
```

A função abaixo irá definir o novo limite para os clientes cujo o Saldo da Conta Poupança ('Account_Balance_Savings') é maior que o valor do limite do Cartão ('Card_Limit').

```python
def generate_new_card_limit(ID):
  return Account_Balance_Savings[ID] if Account_Balance_Savings[ID] > Card_Limit[ID] else Card_Limit[ID]
```

```python
Data_Client['New_Card_Limit'] = list(map(generate_new_card_limit,Data_Client['ID']))
Data_Client
```

Excluindo o Antigo limite do cartão.

```python
Data_Client.drop(columns='Card_Limit', inplace = True)
Data_Client
```
